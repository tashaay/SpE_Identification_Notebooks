```python
import xarray as xr
from datetime import datetime, timedelta


base_path = 'Nc_Files/SMin_contd'
start_datetime_str = "0001-09-29-03600"
end_datetime_str = "0001-10-31-03600"

start_date_str, start_seconds_str = start_datetime_str.split('-03600')
start_datetime = datetime.strptime(start_date_str, "%Y-%m-%d") + timedelta(seconds=3600)

end_date_str, end_seconds_str = end_datetime_str.split('-03600')
end_datetime = datetime.strptime(end_date_str, "%Y-%m-%d") + timedelta(seconds=3600)

file_list = []
current_datetime = start_datetime
date_part_dict = {}  

i=0
while current_datetime <= end_datetime: 
    date_part = current_datetime.strftime("%Y-%m-%d")
    date_part = f"{current_datetime.year:04d}-{current_datetime.month:02d}-{current_datetime.day:02d}"
    seconds_part = "03600"
    file_name = f"SMin_contd.cam.h1.{date_part}-{seconds_part}.nc"
    file_list.append(f"{base_path}/{file_name}")
    date_part_dict[i] = date_part  # Store date_part corresponding to each file
    current_datetime += timedelta(days=1)
    i+=1

for i in range(len(file_list) - 1):
    ds1 = xr.open_dataset(file_list[i])
    ds2 = xr.open_dataset(file_list[i + 1])

    ds_combined = xr.concat([ds1, ds2], dim='time')

    start_time = ds1['time'][-1].values
    end_time = ds2['time'][-2].values

    ds_reformatted = ds_combined.sel(time=slice(start_time, end_time))

    new_file_name = f"SMin_3M_FX2000_f19f19mg16.cam.h1.{date_part_dict[i+1]}-00000.nc"
    print(new_file_name)
    ds_reformatted.to_netcdf(f'{base_path}/{new_file_name}')
```
