```python
season = 'Jun-Aug'

with xr.open_dataset(f'./Nc_Files/SpE_Output/Wuhu_IonTr_run_SpE_Output_{season}.nc', decode_times=False) as ds:
    # Drop the variables you want to remove
    variables_to_remove = ['lev', 'altavg', 'SpEs_Occ_Fr_b_dsavglevLT', 'SpEs_Occ_Fr_b_dsavglev', 
                           'alt_sl_sh_avg_b_dsavg', 'SpEs_Occ_Fr_b_avg', 'SpEs_Occ_Fr_b_dsavgLT', 'alt_sl_sh_avg_b_dsavgLT', 
                           'SpEs_Occ_Fr_b_avgLT', 'SpEs_Occ_Fr_bb_5d_dsavg', 'alt_sl_sh_avg_bb_5d_dsavg', 
                           'SpEsns_Occ_Fr_dsavglev', 'SpEsns_Occ_Fr_avg', 'Mptdens_avv1_b_5d'] 
    ds_modified = ds.drop_vars(variables_to_remove)

    # Change the description of the variable
    ds_modified['SpEs_Occ_Fr_b_dsavg'].attrs['description'] = 'Occ Freq - ds average at each height'
    ds_modified['SpEsns_Occ_Fr_dsavg'].attrs['description'] = 'Occ Freq - ds average at each height'
    ds_modified['SpEs'].attrs['description'] = 'Metal density (144xLT) where Es layer identified (NaNs elsewhere)'
    ds_modified['Mptdens_sh'].attrs['description'] = 'Metal density (144xLT)'
    ds_modified['SpEsns'].attrs['description'] = 'Metal density (lon) where Es layer identified (NaNs elsewhere)'
    ds_modified['Mptdensns'].attrs['description'] = 'Metal density (lon)'
    ds_modified['Mptdens_std'].attrs['description'] = 'standard deviation of the M+ layer in each 2wk time slice'
    ds_modified['Mptdens_avg'].attrs['description'] = 'avg M+ layer for each 2wk time slice'
    
    # Save the modified dataset as a new netCDF file
    ds_modified.to_netcdf(f'./Nc_Files/SpE_Output/Wuhu_IonTr_run_SpE_Output_{season}_David.nc')
```
