# Load WACCM-X data for all seasons and criteria

```python

WACCM_data_alt = '90-150km'
WACCM_UL = 150

run_name = 'Wuhu_IonTr_run'


# Define criteria and corresponding filename appendices
# criteria = ['0.25sigma_2xMpza_1xpeak', '1sigma_3xMpza_1xpeak', '0.5sigma_1xMpza_2xpeak', '0sigma_0.5xMpza_0.5xpeak']

criteria = ['1sigma_3xMpza_1xpeak',  '1sigma_3xMpza_0.5xpeak', '1sigma_2xMpza_1xpeak', '1sigma_2xMpza_0.5xpeak', 
            '0.5sigma_3xMpza_1xpeak', '0.5sigma_2xMpza_1xpeak', '0.5sigma_1xMpza_2xpeak', '0.5sigma_1xMpza_1xpeak', 
            '0.25sigma_2xMpza_1xpeak', '0.25sigma_1xMpza_1xpeak', '0.25sigma_1xMpza_0.5xpeak', 
            '0sigma_2xMpza_1xpeak', '0sigma_1xMpza_1xpeak', '0sigma_0.5xMpza_0.5xpeak']


# Define month folders and corresponding months
month_folders = {
    'Dec-Feb': ['12', '01', '02'],
    'Mar-May': ['03', '04', '05'],
    'Jun-Aug': ['06', '07', '08'],
    'Sep-Nov': ['09', '10', '11']
}

# Initialize empty dictionaries to store datasets
ds_dict = {}
lat_dict = {}
altavg_sl_dict = {}
SpEs_alt_lat_dict = {}
SpEsns_Occ_Freq_lat_dict = {}
tot_SpEs_5deg_dict = {}
SpEs_alt_lt_dict = {}

# Loop through each filename appendix and open respective datasets
for crit in criteria:
    for season in month_folders.keys():
        file_path = f'Nc_Files/SpE_Output/{run_name}_SpE_Output_{season}_{WACCM_data_alt}_{crit}.nc'
        ds_dict[(crit, season)] = xr.open_dataset(file_path, decode_times=False)

        # Store variables with combined keys (criteria, season)
        key_label = (crit, season)

        # Lat line plots
        lat_dict[key_label] = ds_dict[(crit, season)].variables['lat'][:]
        SpEsns_Occ_Freq_lat_dict[key_label] = ds_dict[(crit, season)].variables['SpEsns_Occ_Freq_lat']

        # Alt Lat plots
        altavg_sl_dict[key_label] = ds_dict[(crit, season)].variables['altavg_sl']
        SpEs_alt_lat_dict[key_label] = np.sum(ds_dict[(crit, season)].variables['SpEs_freq_time'], axis=(2, 3, 4))  # 23,96,144,2,3 -> 23,96

        # Alt LT plots
        SpEs_alt_lt_dict[key_label] = np.sum(ds_dict[(crit, season)].variables['SpEs_freq_time'], axis=(3, 4))  # 23,96,144,2,3 -> 23,96,144

        # Rebinning number of measurements into 5deg slices
        bin_edges_5deg = np.arange(-90, 91, 5)
        tot_SpEs_5deg_dict[key_label] = np.zeros((altavg_sl_dict[key_label].shape[0], len(bin_edges_5deg) - 1, 144))

        for i in range(len(bin_edges_5deg) - 1):
            indices_in_bin = np.where((ds_dict[(crit, season)].variables['lat'] >= bin_edges_5deg[i] - 1.89473684) & 
                                      (ds_dict[(crit, season)].variables['lat'] < bin_edges_5deg[i + 1] + 1.89473684))[0]
            fractional_overlap = np.zeros_like(indices_in_bin, dtype=float)

            for j, idx in enumerate(indices_in_bin):
                bin_start = ds_dict[(crit, season)].variables['lat'][idx] - 1.89473684 / 2
                bin_end = ds_dict[(crit, season)].variables['lat'][idx] + 1.89473684 / 2
                overlap_start = max(bin_start, bin_edges_5deg[i])
                overlap_end = min(bin_end, bin_edges_5deg[i + 1])
                overlap_length = max(overlap_end - overlap_start, 0)
                bin_length = bin_end - bin_start
                fractional_overlap[j] = overlap_length / bin_length

            weighted_counts = SpEs_alt_lt_dict[key_label][:, indices_in_bin, :] * fractional_overlap[None, :, None]
            tot_SpEs_5deg_dict[key_label][:, i, :] = np.sum(weighted_counts, axis=1)

```

```python
# Initialize new dictionaries to store normalized data
SpEs_alt_lat_norm_dict = {}

# Define the seasons and criteria
seasons = ['Mar-May', 'Jun-Aug', 'Sep-Nov', 'Dec-Feb']

# Iterate through each criteria
for key in criteria:
    # Gather data for all seasons for this criteria
    seasonal_data = []
    for season in seasons:
        seasonal_data.append( xr.DataArray( SpEs_alt_lat_dict[(key, season)], dims=('lev_sl', 'lat'), coords={'season': season} ) )
    
    # Concatenate along the new 'season' dimension
    combined = xr.concat(seasonal_data, dim='season')
    
    # Calculate the overall min and max values
    overall_min = combined.min(dim=['season', 'lev_sl', 'lat'])
    overall_max = combined.max(dim=['season', 'lev_sl', 'lat'])
    
    # Normalize each season's data using the overall min and max values
    for season in seasons:
        SpEs_alt_lat = xr.DataArray(SpEs_alt_lat_dict[(key, season)], dims=('lev_sl', 'lat'))
        SpEs_alt_lat_norm = (SpEs_alt_lat - overall_min) / (overall_max - overall_min)
        SpEs_alt_lat_norm_dict[(key, season)] = SpEs_alt_lat_norm
```

# Alt - LT

```python
intlat = np.arange(-87.5, 92.5, 5)

# Initialize dictionaries to store normalized data
SpEs_alt_lt_norm_dict = {}

# Define the criteria and seasons
seasons = ['Mar-May', 'Jun-Aug', 'Sep-Nov', 'Dec-Feb']

# Define latitude slice indices
ind_15_20 = np.where((intlat >= 15) & (intlat <= 20))[0]
ind_45_50 = np.where((intlat >= 45) & (intlat <= 50))[0]

# Iterate through each criteria
for key in criteria:
    # Gather data for 15-20 degrees latitude slice for all seasons for this criteria
    seasonal_data_15_20 = []
    seasonal_data_45_50 = []
    
    for season in seasons:
        # Extract data for 15-20 degrees
        SpEs_15_20 = np.squeeze(tot_SpEs_5deg_dict[(key,season)][:, ind_15_20, :])
        SpEs_spr_15_20 = xr.DataArray(SpEs_15_20, dims=('lev_sl', 'LT'), coords={'season': season})
        seasonal_data_15_20.append(SpEs_spr_15_20)
        
        # Extract data for 45-50 degrees
        SpEs_45_50 = np.squeeze(tot_SpEs_5deg_dict[(key,season)][:, ind_45_50, :])
        SpEs_spr_45_50 = xr.DataArray(SpEs_45_50, dims=('lev_sl', 'LT'), coords={'season': season})
        seasonal_data_45_50.append(SpEs_spr_45_50)
    
    # Concatenate along the new 'season' dimension for both latitude slices
    combined_15_20 = xr.concat(seasonal_data_15_20, dim='season')
    combined_45_50 = xr.concat(seasonal_data_45_50, dim='season')
    
    # Calculate the overall min and max values for both latitude slices
    overall_min_15_20 = combined_15_20.min(dim=['season', 'lev_sl', 'LT'])
    overall_max_15_20 = combined_15_20.max(dim=['season', 'lev_sl', 'LT'])
    
    overall_min_45_50 = combined_45_50.min(dim=['season', 'lev_sl', 'LT'])
    overall_max_45_50 = combined_45_50.max(dim=['season', 'lev_sl', 'LT'])
    
    # Normalize each season's data using the overall min and max values for both latitude slices
    for season in seasons:
        # Normalize for 15-20 degrees
        SpEs_altLT_15_20 = xr.DataArray(np.squeeze(tot_SpEs_5deg_dict[(key,season)][:, ind_15_20, :]), dims=('lev_sl', 'LT'))
        SpEs_altLT_15_20_norm = (SpEs_altLT_15_20 - overall_min_15_20) / (overall_max_15_20 - overall_min_15_20)
        SpEs_alt_lt_norm_dict[(key, season, '15_20')] = SpEs_altLT_15_20_norm
        
        # Normalize for 45-50 degrees
        SpEs_altLT_45_50 = xr.DataArray(np.squeeze(tot_SpEs_5deg_dict[(key,season)][:, ind_45_50, :]), dims=('lev_sl', 'LT'))
        SpEs_altLT_45_50_norm = (SpEs_altLT_45_50 - overall_min_45_50) / (overall_max_45_50 - overall_min_45_50)
        SpEs_alt_lt_norm_dict[(key, season, '45_50')] = SpEs_altLT_45_50_norm
```
