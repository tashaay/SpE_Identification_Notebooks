```python
lon = ds.variables['lon']
lat = ds.variables['lat']
lev = ds.variables['lev']

dst = ds.transpose("lev", ...)
Mt = dst.variables[str(Metal)]
Mpt = dst.variables[str(Metal)+'p']
```

```python
ds2t = ds2.transpose("lev", ...)
temp = ds2t.variables['T']
elect = ds2t.variables['e']

geopH = ds2t.variables['Z3'] / 1000 #m-> km
Zavg = geopH.mean(('time','lat', 'lon'))
```

```python
# Slice arrays (lev, altitude) between chosen range

lev_sl_idx_min = 42
lev_sl_idx_max = 66
lev_sl = lev[lev_sl_idx_min:lev_sl_idx_max+1]

lev_sl_ind_min = lev[lev_sl_idx_min] 
lev_sl_ind_max = lev[lev_sl_idx_max] 
lev_sl_str = str("%.1e" % lev_sl_ind_min) + ' hPa : ' + str("%.1e" % lev_sl_ind_max) + ' hPa'
print('Plot lev = ' + lev_sl_str)


Zavg_sl = Zavg[lev_sl_idx_min:lev_sl_idx_max+1]

Zavg_sl_min = Zavg[lev_sl_idx_min] 
Zavg_sl_max = Zavg[lev_sl_idx_max]
Zavg_sl_str = str("%.0f" % Zavg_sl_min) + ' km : ' + str("%.0f" % Zavg_sl_max) + ' km'
print('Approx Z = ' + Zavg_sl_str)

geopH_sl = geopH[lev_sl_idx_min:lev_sl_idx_max+1,:,:,:]
```

```python
# Slice M density and temp by chosen alt range and time range (first 2 weeks of data)

Mpt_sl = Mpt[lev_sl_idx_min:lev_sl_idx_max+1,0:336,:,:]
#Mpt_sl  #lev: 25 time: 336 lat: 96 lon: 144

tempe = temp[lev_sl_idx_min:lev_sl_idx_max+1,0:336,:,:]
```

```python
# VMR to number density calculation

Mptdens = ( Mpt_sl * 1e-6 * 100 * lev_sl ) / (1.380503e-23 * tempe)
```

```python
# Create array with time info for each day
day = np.arange(0,24)
day2 = np.stack((day, day, day, day, day, day, day), axis=0)
day2 = np.reshape(day2, (24*7))   #day2.shape = 168
day3 = np.stack((day2, day2), axis=0)
day3 = np.reshape(day3, (168*2))  #day3.shape = 336

# Create an array with offset needed for each UT time step (24h period)
os =  day * 15 / 2.5  #Offset by 15 degrees lon, lon axis is in 2.5 degree intervals
offset = np.empty(24, dtype=int) #Create offset array so it has same dims as day variable
for d in day:
    offset[d] = int(os[d]) #convert floats to ints

# Stack multiple offset arrays to give offsets for all times in the 2 week dataset 
offset2 = np.stack((offset, offset, offset, offset, offset, offset, offset), axis=0)
offset2 = np.reshape(offset2, (24*7))
offset3 = np.stack((offset2, offset2), axis=0)
offset3 = np.reshape(offset3, (168*2))
#offset3
```

```python
#Create edens_sh array so it has same dims etc as edens variable
#edens_sh = edens
#print(edens) #lev: 25 time: 336 lat: 96 lon: 144 

Mptdens_sh = np.ndarray(Mptdens.shape, dtype = float)


#Iterate over timear, offsetting e density at each time by 15 degrees lon
timear = np.arange(0,336)
for it in timear:
    Mptdens_sh[:,it,:,:] = np.roll(Mptdens[:,it,:,:], offset3[it] , axis=2)
```

```python
# Calculate average of offset e densities
Mptdens_avg = np.mean(Mptdens_sh,  axis=1)

# Calculate std dev of offset e densities
Mptdens_std = np.std(Mptdens_sh, axis=1)
```

```python
# Set offset4 variable to reverse previous roll loop offset
offset4 = - offset3 

# Create M dens_avg_roll & M dens_std_roll arrays so they have same dims etc as edens variable
Mptdens_avg_roll = np.ndarray(Mptdens.shape, dtype = float) 
Mptdens_std_roll = np.ndarray(Mptdens.shape, dtype = float) 

#edens_avg_roll #lev: 25 time: 336 lat: 96 lon: 144

# Iterate over timear, offsetting M density at each time by -15 degrees lon
for it in timear:
    Mptdens_avg_roll[:,it,:,:] = np.roll(Mptdens_avg[:,:,:], offset4[it] , axis=2)
    Mptdens_std_roll[:,it,:,:] = np.roll(Mptdens_std[:,:,:], offset4[it] , axis=2)
```

```python
# Calculate difference between e density from model output and the average e density (rolling with time)
Mptdens_diff = abs(Mptdens-Mptdens_avg_roll)

# Where the difference is less than set value of sigma, set to zero
sigma_val = 1.5   #1sigma~68%   1.5sigma~87%    2sigma~95%
sigma_val_str = str(sigma_val)
SpEs_Mp = Mptdens.where( Mptdens_diff  > (sigma_val*Mptdens_std_roll) , other=0 )

#SpEs #lev: 25 time: 336 lat: 96 lon: 144

#Set string values for sigma symbol
sigma_symb_str = 'sigma: \u03C3'
sigma = sigma_symb_str[-1]
```
