```python
import xarray as xr
import numpy as np
import cftime
import matplotlib.pyplot as plt 
from datetime import datetime 
import matplotlib.colors as colors
import math
%matplotlib inline 

#ds = xr.open_dataset("Nc_Files/Jianfei_WACCMX_files/waccmx_Fe_Fep_Feb.nc")
ds = xr.open_dataset("Nc_Files/Jianfei_WACCMX_files/waccmx_Fe_Fep_Feb.nc", decode_times=False)
ds
```

```python
lev = ds.variables['lev']
lon = ds.variables['lon']
lat = ds.variables['lat']
time = ds.variables['time']
#Fe = ds.variables['Fe']
#Fep = ds.variables['Fep']

dst = ds.transpose("lev", ...)
Fet = dst.variables['Fe']
Fept = dst.variables['Fep']
#'Fet
```

```python
Fe_vert = Fet.mean(('lat', 'lon'))
Fep_vert = Fept.mean(('lat', 'lon'))

Fe_latlev = Fet.mean(('time', 'lon'))
Fep_latlev = Fept.mean(('time', 'lon'))

Fe_lonlev = Fept.mean(('time', 'lat'))
Fep_lonlev = Fept.mean(('time', 'lat'))
```

```python
#Sets up lev / alt axes

def find_nearest(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    return idx

lowest_p = 1*10**-2
highest_p = 5*10**-7

low_ylim_idx = find_nearest(lev, lowest_p)
high_ylim_idx = find_nearest(lev, highest_p)

#print(lev[low_ylim_idx])
#print(lev[high_ylim_idx])

y_lims = [lev[low_ylim_idx], lev[high_ylim_idx]]

lev_plot = lev[high_ylim_idx:low_ylim_idx+1]  #pressure to plot 
print(lev_plot[:])

alt = np.zeros((lev_plot.size)) 
for i in range(lev_plot.size):
  alt[i] = -7.*math.log(lev_plot[i]/1013.)
print(alt[:])

def myround(x, base=5):
    return int(base * round(float(x)/base))

up_alt = myround(alt[0],5)
low_alt = myround(alt[-1],5)

alt_diff=up_alt-low_alt
n =((int(alt_diff) / 5 ) + 1)
n = int(n)
locs=np.linspace(low_alt, up_alt, n, dtype=int)

#locs
```

```python
#__FE__
ticks = np.geomspace(6*10**-13, 2*10**-10, num=20, endpoint=True)

fig, ax = plt.subplots(figsize=(8, 5))
C = ax.contourf(time, lev, Fe_vert, 
                levels=ticks, 
                norm=colors.LogNorm())

ax.set_yscale('log')
ax.set_ylabel('Pressure / HPa')
ax.set_ylim(y_lims[0], y_lims[1])
ax.set_xlabel('Time')
ax2 = ax.twinx()
ax2.set_ylabel('Altitude / km')
ax2.set_yticks(np.linspace(0,10,n))
ax2.set_yticklabels(locs[:])

cbar = fig.colorbar(C,ticks=ticks, pad=0.2)

plt.title('Fe: Time vs Height (Avg Lon & Lat)')
plt.tight_layout()
```
