```python
import xarray as xr
import numpy as np
import cftime
import nc_time_axis
import pandas as pd
import matplotlib.pyplot as plt 
from matplotlib import ticker, cm
from cftime import datetime 
import matplotlib.colors as colors
import math
import random
import matplotlib.cm as mcm
jet = mcm.get_cmap('jet')
import cartopy.crs as ccrs # CRS stands for "Coordinate reference systems" for map projection
from cartopy.crs import PlateCarree
import cartopy.feature as cfeature
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter
from dateutil import tz
import pytz
%matplotlib inline 

filename="Nc_Files/Jianfei_WACCMX_files/waccmx_Fe_Fep_Feb.nc"
#filename="Nc_Files/Jianfei_WACCMX_files/waccmx_Na_Nap_Feb.nc"
#filename="Nc_Files/Jianfei_WACCMX_files/waccmx_Mg_Mgp_Feb.nc"

filename2="Nc_Files/Jianfei_WACCMX_files/waccmx_T_Z3_e_Feb.nc"
#filename3="Nc_Files/Jianfei_WACCMX_files/waccmx_winds_Feb.nc"

M = filename[-10:]
Metal=M[0:2]
#print(Metal)

ds = xr.open_dataset(filename)
ds2 = xr.open_dataset(filename2)

ds2
#ds
```

```python
time = ds.variables['time']
times = xr.cftime_range(start="0001-02-01", periods=672, freq="1H", calendar="noleap")
times2 = ds["time"].dt.strftime("%Y%m%d %H:%M")
times3 = ds["time"].dt.strftime("%Y%m%d")
times4 = ds["time"].dt.strftime("%H:%M")
```

```python
lon = ds.variables['lon']
lat = ds.variables['lat']
lev = ds.variables['lev']

#M = ds.variables[str(Metal)]
#Mp = ds.variables[str(Metal)+'p']

dst = ds.transpose("lev", ...)
Mt = dst.variables[str(Metal)]
Mpt = dst.variables[str(Metal)+'p']
```

```python
ds2t = ds2.transpose("lev", ...)
temp = ds2t.variables['T']
elect = ds2t.variables['e']

geopH = ds2t.variables['Z3'] / 1000 #m-> km
Zavg = geopH.mean(('time','lat', 'lon'))

#Mtdens = ( Mt * 1e-6 * 100 * lev ) / (1.380503e-23 * temp)
Mptdens = ( Mpt * 1e-6 * 100 * lev ) / (1.380503e-23 * temp)
#edens = ( elect * 1e-6 * 100 * lev ) / (1.380503e-23 * temp)
```

```python
#-------------------Plot set up for plot 1-----------------------------
#--------------Mgp density over (12'N lat, 0' lon) as a function of universal time (one day in hrs) and altitude-------------------

lat_plot1_idx = 48
lat_plot1_ind = lat[lat_plot1_idx] #
lat_plot1_str = str("%.1f" % lat_plot1_ind)
print('Plot lat = ' + lat_plot1_str)

lon_plot1_idx = 72
lon_plot1_ind = lon[lon_plot1_idx] 
lon_plot1_str = str("%.1f" % lon_plot1_ind)
print('Plot lon = ' + lon_plot1_str)

lev_plot1_idx_min = 19
lev_plot1_idx_max = 72
lev_plot1_ind_min = lev[lev_plot1_idx_min] 
lev_plot1_ind_max = lev[lev_plot1_idx_max] 
lev_plot1 = lev[lev_plot1_idx_min:lev_plot1_idx_max+1]
lev_plot1_str = str("%.1e" % lev_plot1_ind_min) + ' hPa : ' + str("%.1e" % lev_plot1_ind_max) + ' hPa'
print('Plot lev = ' + lev_plot1_str)

tim_plot1_idx = 1
tim_plot1_ind = times2[tim_plot1_idx] 

tim_plot1_inds = times[tim_plot1_idx] 
tim_plot1_str = str(tim_plot1_inds) 


Z_plot1_ind_min = geopH[lev_plot1_idx_min,tim_plot1_idx,lat_plot1_idx,lon_plot1_idx] #lev: 126 time: 672 lat: 96 lon: 144
Z_plot1_ind_max = geopH[lev_plot1_idx_max,tim_plot1_idx,lat_plot1_idx,lon_plot1_idx]
Z1_plot1 = geopH[lev_plot1_idx_min:lev_plot1_idx_max+1,tim_plot1_idx,lat_plot1_idx,lon_plot1_idx]
lev_plot1_str = str("%.0f" % Z_plot1_ind_min) + ' km : ' + str("%.0f" % Z_plot1_ind_max) + ' km'
print('Approx Z = ' + lev_plot1_str)

print('Plot time (UT) = ' + tim_plot1_str)

Mptplot1 = Mptdens[lev_plot1_idx_min:lev_plot1_idx_max+1,tim_plot1_idx,lat_plot1_idx,lon_plot1_idx] 
```

```python
#Calc local time coversion

if lon_plot1_ind > 180.:
    lon_plot1 =  lon_plot1_ind - 360
else:
    lon_plot1 = lon_plot1_ind

#print(lon_plot1)

timeconv = lon_plot1 / 15
timeconvr = timeconv.round()
timeconvrr = int(timeconvr)
print('+' + str(timeconvrr)+'h')

timeUTplot1 = int(tim_plot1_str[-8:-6])
print('time (UT) = ' + str(timeUTplot1))

timeLTplot1 = timeUTplot1 + timeconvrr
print('time (LT) = ' + str(timeLTplot1))
```

```python
fig, ax1 = plt.subplots(figsize=(6, 4))

ax1.plot(Mptplot1, lev_plot1, label='Fe+', color='black', linestyle='--')

#ax1.set_xscale('log')

#ax1.set_ylim(y_lims[0], y_lims[1])
ax1.set_yscale('log')
ax1.set_ylabel('Pressure / HPa')

#ax1.legend()

#ax1.set_xlim(10**-10, 10**9)
#ax1.set_xlabel(r'Density / cm$^{-3}$')

#--------------------------------------------

#set up second axis: altitude
ax2 = ax1.twinx()
ax2.set_yticks(Z1_plot1)
#ax2.set_yticklabels(locs[:])
#ax2.set_ylabel('Altitude / km')

#for label in ax2.set_yticklabels(locs[:])[::2]:
#    label.set_visible(False)
#--------------------------------------------

plt.tight_layout()
figname='Fe_alt.png'
#plt.savefig('./Figures/Jianfei_run/'+str(figname))

plt.show()
```
